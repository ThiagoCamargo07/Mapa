```python
import folium
from IPython.display import display

map_center = [10.0760, 72.8777]
mymap = folium.Map(location=map_center, zoom_start=10)


folium.Marker(
    [10.0760, 21.877],
    popup="Mumbai",
    icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(mymap)

display(mymap)
```
